```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt # Matplotlib
%matplotlib inline
```

```python
df = pd.read_csv('California_airbnb.csv')
df = df.dropna()
```

```python
df
```

```python
# Figure of data histogram by region
plt.figure(figsize=(10, 4), dpi = 100)

plt.hist(df['Region'], bins = 25, color = 'blue', edgecolor = 'gray')

plt.xlabel('Regions', fontsize = 14)
plt.ylabel('# of Datapoints', fontsize = 14)

plt.title('Distribution of Datapoints in each Los Angeles Region', fontsize = 14)

plt.xticks(fontsize = 8, rotation = 45)
plt.yticks(fontsize = 8);
```

```python
# Figure of data histogram by price
prices = df[df['price'] < 1000]

plt.figure(figsize=(10, 4), dpi = 100)

plt.hist(prices['price'], bins = 40, color = 'brown', edgecolor = 'gray')

plt.xlabel('Price/Night ($)', fontsize = 14)
plt.ylabel('# of Datapoints', fontsize = 14)

plt.title('Distribution of Datapoints by Airbnb Price/Night', fontsize = 14)

plt.xticks(fontsize = 8)
plt.yticks(fontsize = 8);
```

```python
# Bar plot
plt.figure(figsize=(8, 4))
plt.barh(df['property_type'].unique(), df['property_type'].value_counts(), color = ['orange', 'brown'])

plt.title('Number of Datapoints of each Property Type', fontsize = 14)
plt.xlabel("# of Datapoints", fontsize = 12)
plt.ylabel("Property Types", fontsize = 12)
plt.yticks(fontsize = 7)
plt.xticks(fontsize = 10); 
```
